#Parte 1. Base de datos y almacen de datos

Una base de datos se define como un conjunto de datos almacenados que pueden guardarse en una computadora o en la nube para poder acceder a ellos. Existen diferentes  bases de datos, algunos de ellos pueden ser relacionales y los no relacionales. Las bases de datos relacionales están más estructuradas y organizadas que suelen ser en SQL, y las no relacionales suelen estar por clases. Entre ellas también se conocen como ETL (extraer, transformar y cargar) o ELT (extraer, cargar y transformar). La primera es más fácil y mejor por su estructura más controlada por las organizaciones, pero la segunda viene más de datos externos. Ambas bases de datos son importantes las cuales se tiene que ordenar para tener mayor análisis de datos. 


Un almacén de datos es un sistema de gestión de datos que permite habilitar y permitir actividades, el cual están destinados para consulta y análisis de grandes fuentes de datos históricos. También centraliza y consolida grandes cantidades de datos de múltiples fuentes que ayudan en la toma de decisiones. Los datos se organizan dentro de tablas o columnas las cuales cada columna contiene datos para después analizarlas y visualizarlas. Los almacenes de datos también puede ser desde la nube como de forma local, esto para facilitar su accesibilidad en ellas las cuales son muy útiles para los usuarios.
